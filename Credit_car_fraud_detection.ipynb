```python
import numpy as np
import pandas as pd
import joblib as jb
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
card_data = pd.read_csv('creditcard.csv')
```

```python
card_data.head()
```

```python
card_data.tail()
```

```python
card_data.info()
```

```python
card_data.isnull().sum()
```

```python
card_data.shape
```

```python
card_data['Class'].value_counts()
```

```python
legit = card_data[card_data.Class == 0]
fraud = card_data[card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
card_data.groupby('Class').mean()
```

```python
legit_sample = legit.sample(n=492)
```

```python
new_dataset = pd.concat([legit_sample , fraud ], axis =0)
```

```python
new_dataset.shape
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
X = new_dataset.drop(columns = 'Class' , axis = 1)
Y = new_dataset['Class']
```

```python
print(X.shape)
print(Y.shape)
```

```python
print(X)
print(Y)
```

```python
X_train , X_test , Y_train , Y_test = train_test_split(X , Y , test_size = 0.2 , stratify = Y , random_state = 2)bb
```

```python
print(X.shape , X_train.shape , X_test.shape)
print(Y.shape , Y_train.shape , Y_test.shape)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train , Y_train)
```

```python
train_predictions = model.predict(X_train)
test_predictions = model.predict(X_test)

train_score = accuracy_score(train_predictions , Y_train)

test_score = accuracy_score(test_predictions , Y_test)

print('The accuracy score on training data is : ', train_score)

print('The accuracy score on test data is : ', test_score)
```

```python
jb.dump(model ,'CREDIT_CARD_FRAUD_DETECTION.joblib')
```
